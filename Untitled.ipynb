```python
s = jumpman_data.isna().sum()
s = s.iloc[s.to_numpy().nonzero()]

labels = s.index
ax = s.sort_values().plot.barh(figsize=(8,4))

for rect in ax.patches:
    y_value = rect.get_y() + rect.get_height() / 2
    x_value = rect.get_width()
    ax.annotate(str(x_value)+' '+'('+str(round(x_value/len(jumpman_data)*100, 2)) + '%)',
                (x_value, y_value),
                va='center',
                textcoords="offset points",
                xytext=(2,0),
                fontsize=12)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.title('Null Values',fontsize=16)
plt.show()
```

```python
item_name_place_category_both_null = jumpman_data[jumpman_data['item_name'].isna()]['place_category'].isna().value_counts()[True]
item_name_place_how_long_null = jumpman_data[jumpman_data['item_name'].isna()]['how_long_it_took_to_order'].isna().value_counts()[True]

print(item_name_place_category_both_null/s.place_category*100)
print(item_name_place_how_long_null/s.how_long_it_took_to_order*100)
```

```python
list_df = jumpman_data['how_long_it_took_to_order'].dropna().str.split(':')
time_df = pd.DataFrame(list_df.values.tolist(), columns=['Hrs','Mins','Secs'])
time_df.head()
```

```python
time_df['Mins'] = time_df['Mins'].astype(int)
minutes_list = sorted(list(time_df['Mins'].values))
minutes_mean = time_df['Mins'].describe().loc['mean']
minutes_std = time_df['Mins'].describe().loc['std']

plt.figure(figsize=(12,6))
plt.hist(minutes_list, bins=60)
plt.xticks(np.arange(0,60,2))
plt.axvline(minutes_mean, color='black')
plt.axvline(minutes_mean+minutes_std, color='black', linestyle='--')
plt.axvline(minutes_mean-minutes_std, color='black', linestyle='--')
plt.axvline(minutes_mean+(minutes_std*2), color='black', linestyle='-.')
plt.axvline(minutes_mean+(minutes_std*3), color='black', linestyle=':')
plt.xlabel('Time to Place Order (Min)',fontsize=14)
plt.ylabel('Count of Orders',fontsize=14)
plt.title('Time to Place Order Distribution',fontsize=16)
plt.show()

print(time_df['Mins'].describe())
```

```python
T = (time_df['Mins']>minutes_mean+(minutes_std*3)).value_counts()[True]
F = (time_df['Mins']>minutes_mean+(minutes_std*3)).value_counts()[False]

(T/F)*100
```

1.9% of data with **how_long_it_took_to_order values** values took over 23 minutes to order.

```python
ind = time_df[time_df.Mins > minutes_mean+(minutes_std*3)].index
long_order_times = jumpman_data.dropna(axis=0,how='any',subset=['how_long_it_took_to_order']).reset_index(drop=True).loc[ind]
long_order_times['pickup_place'].value_counts().head()
```

```python
time_format = '%Y-%m-%d %H:%M:%S.%f'

hour = [datetime.strptime(time,time_format).hour for time in jumpman_data['when_the_delivery_started'].dropna(axis=0,how='any')]
s = pd.Series(hour).describe()


plt.figure(figsize=(8,5))
plt.hist(hour,bins=24)
plt.xticks(np.arange(0,24,2), fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0,23)
plt.axvline(s['mean'],color='black')
plt.axvline(s['mean']+s['std'],color='black',linestyle='--')
plt.axvline(s['mean']-s['std'],color='black',linestyle='--')
plt.xlabel('Hour of Day',fontsize=14)
plt.ylabel('Count of Orders',fontsize=14)
plt.title('Total Orders by Hour Over 30 Days', fontsize=16)
plt.show()
print(s)
```

```python
day = [datetime.strptime(time,time_format).day for time in jumpman_data['when_the_delivery_started'].dropna(axis=0,how='any')]
day_df = pd.DataFrame({'Day':day, 'Daily Total':1}).groupby('Day').sum()

five_day_mean = []
j=0
for i in np.arange(5,31,5):
    five_day = day_df.loc[j:i]
    j=i+1
    five_day_mean.append(five_day.mean()['Daily Total'])
five_day_df = pd.DataFrame({'Day':np.arange(5,31,5), '5 Day Mean':five_day_mean}).set_index('Day')


ax = day_df.plot(figsize=(8,6))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Day of Month', fontsize=14)
plt.ylabel('Count of Orders',fontsize=14)

five_day_df.plot(color='orange',marker='o',ax=ax)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Day of Month', fontsize=14)
plt.ylabel('Count of Orders',fontsize=14)
plt.title('Orders Over Last 30 Days',fontsize=16)

plt.show()
```

```python
pickup_coords = jumpman_data[['pickup_lat','pickup_lon']]
dropoff_coords = jumpman_data[['dropoff_lat','dropoff_lon']]

distance = []
for i in range(len(jumpman_data)):
    row = jumpman_data.loc[i]
    pickup_coords = (row['pickup_lat'],row['pickup_lon'])
    dropoff_coords = (row['dropoff_lat'],row['dropoff_lon'])
    distance.append(geopy.distance.distance(pickup_coords, dropoff_coords).miles)
    
jumpman_data['delivery_distance'] = distance
jumpman_data.head()
```

```python
delivery_time_df = jumpman_data.copy().dropna(axis=0,how='any',subset=['when_the_Jumpman_left_pickup'])
wait = []

for i in delivery_time_df.index:
    time_ = delivery_time_df.when_the_Jumpman_arrived_at_dropoff.loc[i]
    time_2 = delivery_time_df.when_the_Jumpman_left_pickup.loc[i]
    wait_time = datetime.strptime(time_,time_format) - datetime.strptime(time_2,time_format)
    wait.append(wait_time.seconds/60)
    
delivery_time_df['delivery_time'] = wait
delivery_time_df.head()
```

```python
delivery_time_df['delivery_distance'].corr(delivery_time_df['delivery_time'])
```

```python
miles = []
mean_time = []

distance_time_df = delivery_time_df[['delivery_distance','delivery_time']]
for i in range(int(round(distance_time_df['delivery_distance'].max()))):
    j = i+1
    mean_time.append(distance_time_df[(distance_time_df['delivery_distance'] > i) & (distance_time_df['delivery_distance'] < j)]['delivery_time'].mean())
    miles.append(str(i)+'-'+str(j))
    
distance_interval_df = pd.DataFrame({'Miles':miles,'Mean Mins':mean_time}).set_index('Miles')

distance_interval_df.plot(marker='o',figsize=(8,6),legend=False)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Miles',fontsize=14)
plt.ylabel('Minutes',fontsize=14)
plt.title('Average Delivery Time',fontsize=16)
plt.show()
```
